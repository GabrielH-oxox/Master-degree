```python
import pandas as pd
import glob
```

```python
ssp_values = ["ssp126", "ssp245", "ssp370", "ssp585"]
esm_values = ["ACCESS-CM2", "ACCESS-ESM1-5", "AWI-CM-1-1-MR", "BCC-CSM2-MR", "CanESM5", "CMCC-ESM2", "EC-Earth3", "EC-Earth3-Veg", 
              "EC-Earth3-Veg-LR", "FGOALS-g3", "GFDL-ESM4", "IITM-ESM", "INM-CM4-8", "INM-CM5-0", "IPSL-CM6A-LR", "KACE-1-0-G",
              "KIOST-ESM", "MIROC6", "MPI-ESM1-2-HR", "MPI-ESM1-2-LR", "MRI-ESM2-0", "NESM3", "NorESM2-LM", "NorESM2-MM", "TaiESM1"]

output_path = r"C:\Users\USER\Desktop\MA\SEM Lab\碩論內容\AR6資料\濁水溪\整理後\最低溫"
# 讀取並處理資料
for ssp_value in ssp_values:
    for esm_value in esm_values:
        data_path = rf"C:\Users\USER\Desktop\MA\SEM Lab\碩論內容\AR6資料\濁水溪\整理前\最低溫\{ssp_value}"
        file_pattern = f"{data_path}\\AR6_統計降尺度_日資料_濁水溪_最低溫_{ssp_value}_{esm_value}_*.csv"
        print(f"Searching for: {file_pattern}")
        files = glob.glob(file_pattern)
        dataframes = []
        for i, file in enumerate(files):
            df = pd.read_csv(file)
            df = df.T[2:-1]  # 轉置資料並刪除不需要的列
            new_columns = [f"J-{i}" for i in range(1, len(df.columns) + 1)]  # 創建索引列表
            df.columns = new_columns  # 設定新索引
            dataframes.append(df)

        # 合併資料
        if dataframes:
            merged_df = pd.concat(dataframes, axis=0, ignore_index=False)

            # 輸出合併後的資料
            output_file = f"{output_path}\\{ssp_value}_{esm_value}_最低溫整合.csv"
            merged_df.to_csv(output_file, header=1)
        else:
            print(f"No matching files found for {ssp_value}, {esm_value}")
```
